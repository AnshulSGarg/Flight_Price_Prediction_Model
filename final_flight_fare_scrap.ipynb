```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import pandas as pd
import time
import datetime
import random
```

```python
def carrier(soup):
        #carrier
    carrier = []
    k = 1
    container = soup.find_all('div',{'class':'sSHqwe tPgKwe ogfYpf'})
    for i in container:
        k = k + 1
        try:
            carrier.append(i.find('span').text)
        except:
            carrier.append('error')
    flight_carrier = []
    j = 0
    for j in range(round(len(carrier)/2)):
        p = 2*j
        # print(carrier[p])    
        flight_carrier.append(carrier[p])
    return flight_carrier
```

```python
def layover(soup):
        #layover
    layover_info = []
    container = soup.find_all('div', {'class': 'sSHqwe tPgKwe ogfYpf'})
    for i in container:
            aria_label = i.get('aria-label')
            if aria_label:
                layover_info.append(aria_label)
            else:
                layover_info.append('')
    layover = []
    j = 0
    for j in range(round(len(layover_info)/2)):
        p = 2*j + 1
        # print(carrier[p])    
        layover.append(layover_info[p])
    return layover
```

```python
def flight_time(soup):
    # flight time
    flight_time = []
    k = 1
    container = soup.find_all('div',{'class':'zxVSec YMlIz tPgKwe ogfYpf'})
    for i in container:
        k = k + 1
        try:
            flight_time.append(i.find('span').text)
        except:
            flight_time.append('error')
    return flight_time
```

```python
def from_loc(soup):    
#from loc
    from_loc = []
    k = 1
    container = soup.find_all('div',{'class':'G2WY5c'})
    for i in container:
        k = k + 1
        try:
            from_loc.append(i.find('div').text)
        except:
            from_loc.append('error')
    return from_loc
```

```python
def to_loc(soup):    
#to loc
    to_loc = []
    k = 1
    container = soup.find_all('div',{'class':'c8rWCd'})
    for i in container:
        k = k + 1
        try:
            to_loc.append(i.find('div').text)
        except:
            to_loc.append('error')
    return to_loc
```

```python
def stop(soup):    
#stop
    stop = []
    k = 1
    container = soup.find_all('div',{'class':'hF6lYb'})
    for i in container:
        k = k + 1
        try:
            stop.append(i.find('span',{'class':'VG3hNb'}).text)
        except:
            stop.append('error')
    return stop
```

```python
def carbon(soup):
    carbon = []

    div_elements = soup.find_all("div", class_="NZRfve")

    # Loop through each div element
    for div in div_elements:
        # Extract aria-label text
        aria_label = div.get("aria-label")
        if aria_label:
            emission_estimate = aria_label
            # print("Emission Estimate:", emission_estimate)
            carbon.append(emission_estimate)
        else:
            carbon.append("Unknown")
            carbon.append("Unknown")
            carbon.append("Unknown")
            # print("Emission Estimate: Unknown")
    # print(carbon)
    carbon_emission = []
    j = 0
    for j in range(round(len(carbon)/3)):
        p = 3*j
        # print(carrier[p])    
        carbon_emission.append(carbon[p])
    return carbon_emission



```

```python
def bag(soup):    
#overhead bin
    bag = []
    k = 1
    container = soup.find_all('div',{'class':'BVAVmf'})
    for i in container:
        # print(i)
        k = k + 1
        try:
            bag.append(i.get_text())
        except:
            bag.append('error')
    overhead_bin = []
    j = 0
    for j in range(round(len(bag)/3)):
        p = 3*j
        # print(carrier[p])    
        overhead_bin.append(bag[p])
    return overhead_bin
```

```python
def details(soup):    
    details = []
    k = 1
    container = soup.find_all('div',{'class':'JMc5Xc'})
    for i in container:
        k = k + 1
        try:
            details.append(i.get('aria-label'))
        except:
            details.append('error')
    return details
```

```python
def google_search(driver,d):
    time.sleep(1)
    from_input = driver.find_element(by=By.XPATH, value='/html/body/div[1]/div[3]/form/div[1]/div[1]/div[1]/div/div[2]/textarea')
    from_input.clear()
    time.sleep(1)
    from_input.send_keys(d)
    time.sleep(1)
    from_input.send_keys(Keys.ENTER)
    time.sleep(2)
```

```python
def scrap(driver,d):
    driver.get("https://www.google.com")
    google_search(driver,d)
    try:
        next_click(driver)
    except:
        pass
    html = driver.page_source
    soup = BeautifulSoup(html,'lxml')
    
    carrier_list = carrier(soup)
    layover_list = layover(soup)
    flight_time_list = flight_time(soup)
    from_loc_list = from_loc(soup)
    to_loc_list = to_loc(soup)
    stop_list = stop(soup)
    carbon_list = carbon(soup)
    bag_list = bag(soup)
    details_list = details(soup)

    time.sleep(1)
    driver.quit()
    time.sleep(1)

    # if carrier_list == carbon_list:
    #     pass
    # else:
    #     carbon_list = carrier_list

    df = pd.DataFrame({
        'carrier':carrier_list,
        # 'price':price,
        'layover':layover_list,
        'flight_time':flight_time_list,
        'from_loc':from_loc_list,
        'to_loc':to_loc_list,
        'stop':stop_list,
        'overhead_bin':bag_list,
        'carbon_emission':carbon_list,
        'details':details_list})


    df['details'] = df['details'].astype(str)
    df['len'] = df['details'].apply(lambda x: len(x.split('.')))
    import re
    df['price_1'] = df['details'].apply(lambda x:x.split('.')[0])
    def extract_numbers(text):
        numbers = re.findall(r'\d+', text)
        if numbers:
            return int(numbers[0])
        else:
            return None
    df['price'] = df['price_1'].apply(extract_numbers)
    df['from_time'] = df['flight_time'].apply(lambda x: x.split("–")[0].strip())
    df['to_time'] = df['flight_time'].apply(lambda x: x.split("–")[1].strip())
    def extract_characters(text):
        if len(text) >= 7 and text[6] == 'M':
            return text[:7]
        elif len(text) >= 8 and text[7] == 'M':
            return text[:8]
        else:
            return 'Error'  # Or any other error handling you prefer   
    df['from_timestamp'] = df['from_time'].apply(extract_characters)
    df['to_timestamp'] = df['to_time'].apply(extract_characters)
    df['from_date'] = df['from_time'].str.split(', ').values[0][1]
    df['to_date'] = df['to_time'].str.split(', ').values[0][1]
    final_df = df[['carrier', 'from_loc', 'to_loc', 'stop','price','from_timestamp', 'to_timestamp', 'from_date', 'to_date','carbon_emission','overhead_bin','layover', 'details']]
    final_df['Report_Run_Time'] = pd.Timestamp.now().strftime('%Y-%m-%d')
    final_df = final_df[[ 'Report_Run_Time','carrier', 'from_loc', 'to_loc', 'stop', 'price', 'from_timestamp','to_timestamp', 'from_date', 'to_date', 'carbon_emission', 'overhead_bin', 'layover', 'details']]
    return final_df


```
